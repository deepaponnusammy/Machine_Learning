```python
#Factorial Function
def Factorial_value(num):
  Factorial = 1
  if num == 0:
    return Factorial
  else:
    for i in range(1, num +1):
      Factorial = Factorial * i
    return Factorial
print(Factorial_value(3))
print(Factorial_value(6))
print(Factorial_value(9))
```

```python
#adding title to the x & y axis label
#sine wave
plt.plot(x,y)
plt.xlabel('Angle')
plt.ylabel('Sine Value')
plt.title('Sine Wave')
plt.show()
```

```python
#Bar plot
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['English', 'French', 'Spanish', 'Latin', 'German']
people = [100, 50, 150, 40, 70]
ax.bar(languages, people)
plt.xlabel('LANGUAGES')
plt.ylabel('NUMBER OF PEOPLE')
plt.show()
```

```python
#Pie chart
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['English', 'French', 'Spanish', 'Latin', 'German']
people = [100, 50, 150, 40, 70]
ax.pie(people, labels = languages, autopct = '%1.1f%%')
plt.show()
```

```python
#Scatter plot
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
plt.scatter(x, y, color ='g')
plt.scatter(x, z, color ='b')
plt.show()
```

```python
#3D Scatter plot
fig1 = plt.figure()
ax = plt.axes(projection = '3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x, y, z, c = z, cmap ='Blues')
plt.show()
```

```python
#house price dataset
from sklearn.datasets import load_boston
house_boston = load_boston()
house = pd.DataFrame(house_boston.data, columns = house_boston.feature_names)
house['PRICE'] = house_boston.target
house.head()
```

```python
#Distribution plot
sns.distplot(house['PRICE'])
```

```python
sns.displot(house['PRICE'])
```

```python
#correlation
#1. Positive correlation(one value increase another value increase)
#2. Negative correlation(one value increase another value decrease)
#Heat Map
correlation = house.corr()
```

```python
#constructing a heat map
plt.figure(figsize = (10, 10))
sns.heatmap(correlation, cbar = True, square = True, fmt = '.1f', annot = True, annot_kws = {'size' : 8}, cmap = 'Blues')
```
